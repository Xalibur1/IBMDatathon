# The Bar graphs of the non numeric data

```python
analysis_df = df
score_list = df['Exam_Score'].tolist()
```

<h3>The following code divides into the ranges of scores:</h3>

```python
for i in range(len(score_list)):
    if score_list[i] == 100:
        score_list[i] = 100
    elif score_list[i] >= 90:
        score_list[i] = 90
    elif 80 <= score_list[i] < 90:
        score_list[i] = 80
    elif 70 <= score_list[i] < 80:
        score_list[i] = 70
    elif 60 <= score_list[i] < 70:
        score_list[i] = 60
    elif 50 <= score_list[i] < 60:
        score_list[i] = 50
    elif 40 <= score_list[i] < 50:
        score_list[i] = 40
    elif 30 <= score_list[i] < 40:
        score_list[i] = 30
    elif 20 <= score_list[i] < 30:
        score_list[i] = 20
    elif 10 <= score_list[i] < 20:
        score_list[i] = 10
    elif 0 <= score_list[i] < 10:
        score_list[i] = 0
analysis_df['approx % (>=)'] = score_list
df['approx'] = score_list
```

```python
medium_parental = df[df['Parental_Involvement'] == 'Medium'].head(1200)
high_parental = df[df['Parental_Involvement'] == 'High'].head(1200)
low_parental = df[df['Parental_Involvement'] == 'Low'].head(1200)

colors = ['#FF6347', '#4682B4', '#32CD32', '#FFD700', '#8A2BE2'] 

fig, axes = plt.subplots(1, 3, figsize=(18, 6))


pd.crosstab(medium_parental['Parental_Involvement'], medium_parental['approx']).plot(kind='bar', ax=axes[0], color=colors)
axes[0].set_title('Medium Parental Involvement')

pd.crosstab(high_parental['Parental_Involvement'], high_parental['approx']).plot(kind='bar', ax=axes[1], color=colors)
axes[1].set_title('High Parental Involvement')

pd.crosstab(low_parental['Parental_Involvement'], low_parental['approx']).plot(kind='bar', ax=axes[2], color=colors)
axes[2].set_title('Low Parental Involvement')

plt.tight_layout()
plt.show()
```

```python
low_access = df[df['Access_to_Resources'] == 'Low'].head(300)
medium_access = df[df['Access_to_Resources'] == 'Medium'].head(300)
high_access = df[df['Access_to_Resources'] == 'High'].head(300)

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

pd.crosstab(low_access['Access_to_Resources'], low_access['approx']).plot(kind='bar', ax=axes[0], color=colors)
axes[0].set_title('Low Access to Resources')

pd.crosstab(medium_access['Access_to_Resources'], medium_access['approx']).plot(kind='bar', ax=axes[1], color=colors)
axes[1].set_title('Medium Access to Resources')

pd.crosstab(high_access['Access_to_Resources'], high_access['approx']).plot(kind='bar', ax=axes[2], color=colors)
axes[2].set_title('High Access to Resources')

plt.tight_layout()
plt.show()
```

```python
ea_yes = df[df['Extracurricular_Activities'] == 'Yes'].head(600)
ea_no = df[df['Extracurricular_Activities'] == 'No'].head(600)

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

pd.crosstab(ea_yes['Extracurricular_Activities'], ea_yes['approx']).plot(kind='bar', ax=axes[0], color=colors)
axes[0].set_title('With Extracurricular Activities')

pd.crosstab(ea_no['Extracurricular_Activities'], ea_no['approx']).plot(kind='bar', ax=axes[1], color=colors)
axes[1].set_title('Without Extracurricular Activities')

plt.tight_layout()
plt.show()
```

```python
nety = df[df['Internet_Access'] == 'Yes'].head(300)
netn = df[df['Internet_Access'] == 'No'].head(300)

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

pd.crosstab(nety['Internet_Access'], nety['approx']).plot(kind='bar', ax=axes[0], color=colors)
axes[0].set_title('With Internet Access')

pd.crosstab(netn['Internet_Access'], netn['approx']).plot(kind='bar', ax=axes[1], color=colors)
axes[1].set_title('Without Internet Access')

plt.tight_layout()
plt.show()
```

<h1> Key findings </h1>

<h3>Parental Involvement:</h3> Students with high parental involvement consistently demonstrated better academic performance.

<h3>Access to Resources:</h3> While students with high access to resources tended to perform well, it's notable that some students with medium or low access achieved exceptional results.

<h3>Extracurricular Activities:</h3> Participation in extracurricular activities was associated with higher overall performance, although some students without involvement also excelled.

<h3>Internet Access:</h3> Surprisingly, students without internet access often outperformed those with internet access, suggesting that internet access may not be a decisive factor.

<h3>Motivation Levels:</h3> High motivation was generally correlated with better performance, but individual variations were observed.

<h3>Family Income:</h3> Students from higher-income families tended to have better academic outcomes.

<h3>Teacher Quality:</h3> The quality of teachers significantly impacted student performance.

<h3>School Type:</h3> While private schools often yielded better results, this factor should be interpreted with caution due to potential confounding variables.

<h3>Peer Influence:</h3> Positive peer influence was associated with improved academic performance.

<h3>Learning Disabilities:</h3> Students without learning disabilities generally outperformed those with disabilities.

<h3>Parental Education:</h3> Parental education level was not a significant predictor of student performance.

<h3>Distance from Home:</h3> Proximity to campus was correlated with better performance.

<h2>Conclusion</h2>

Based on the data analyzed, parental involvement, access to resources, teacher quality, peer influence, and motivation levels emerged as key factors influencing student performance. While family income, school type, and distance from home were also relevant, their effects were more nuanced. It's important to note that individual variations exist, and other factors not considered in this analysis may also play a role.

<h2>Recommendations</h2>

Prioritize Parental Involvement: Schools should foster strong partnerships with parents to support student learning.

Ensure Equitable Access: Efforts should be made to provide all students with access to necessary resources.
    
Promote Positive Peer Culture: Create a supportive and inclusive school environment that encourages positive peer interactions.
    
Invest in Teacher Development: Continuously invest in professional development for teachers to enhance their skills and effectiveness.

Address Learning Disabilities: Provide appropriate accommodations and support for students with learning disabilities.
    
By addressing these key factors, schools can create a more conducive learning environment and help students achieve their full potential.
